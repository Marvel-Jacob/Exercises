### •	Display the ratio of male vs female cast members per movie in current order 

```python
data['cast'][0][0]['gender']
```

```python
data['ratio_in_cast']=0
```

```python
c_male=0
c_female=0
for i in range(len(data)):
    for j in range(len(data['cast'][i])):
        if data['cast'][i][j]['gender']==1:
            c_female=c_female+1
        else:
            c_male=c_male+1
    ratio=c_male/c_female  
    data['ratio_in_cast'][i]=ratio
```

### •	Display the total count of movies acted in per cast member in any order.
