```python
import pandas as pd
import numpy as np
```

```python
air=pd.read_csv('C:/Users/Administrator/Python DEMOS/CASE STUDIES/airquality.csv')
```

```python
#1.	Fetch the observations for 9 day of June
air.query("Month == 6 & Day == 9")

#2.	Find Average temperature for the month of June. 
air.query("Month == 6").Temp.mean()

#3.	To which day of June has the least temperature
air_missing = air.dropna()
air_missing
air_missing.columns
air_missing.shape
ab= air_missing.query("Month == 6" ).Temp.min()
air_missing[(air_missing.Temp == ab)& (air_missing.Month == 6)].Day

#isnull is a function which tells us whether a column has null or not '"~"symbol is used for NOT'

#4.	Find Maximum Ozone value for the month of May
air_missing.query("Month == 5").Ozone.max()

#5.	Find the count of the missing values in the ozone column of the data set
air[(air.Ozone.isnull())].Ozone.shape

#6.	Find out What is the mean of the Ozone column in this dataset
air_missing.Ozone.mean()

#7.	Find out which month has the highest temperature
air_missing[air_missing.Temp == air_missing.Temp.max()].Month

#8.	Find out the wind value when the Ozone becomes maximum
air_missing[air_missing.Ozone == air_missing.Ozone.max()].Wind

#9.	Find out the months for which the airquality observations have been carried out
air[~air.Month.isnull()].Month.unique()

#10.	Find the Ozone and temperature values for the 1st observation of every month.
air.query("Day == 1")[["Ozone","Temp"]]

#11.	Which day of which month corresponds to the least Ozone Value. 
air_missing[air_missing.Ozone == air_missing.Ozone.min()].Month

#12.	Convert the temperature for all the observations to Centigrade scale
air.Temp.apply(lambda Temp: (Temp - 32)*5/9)
```
